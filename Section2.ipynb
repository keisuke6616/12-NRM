```
import numpy as np

import numpy.linalg as la
import matplotlib.pyplot as plt

import time
```

# Figure 1

```
t1 = time.time()


N = 2
Times = 20
dim = [2, 20, 200, 2000]
N_dim = len(dim)
index = [i for i in range(N_dim)]


result = np.zeros((N_dim, Times, N, N))

for p, pc in zip(dim, index):
    print('Dimension : {}'.format(p))
    print('Calcurating...')
    
    for t in range(Times):
        X = np.random.normal(0, 1, (N, p))
        Sd = np.dot(X, X.T) / N
        
        value, vec = la.eig(Sd)
        for n in range(N):
            result[pc, t, n] = float(N) * value[n] * vec.T[n] / float(p)
        
    print('Result')
    print(result[pc])
    print('Finished')
    print('')

t2 = time.time() - t1
print('Calcuration time : {}'.format(t2))
```

```
plt.figure(figsize=(10, 10))
title_list = ['dim = {}'.format(p) for p in dim]
for i in range(N_dim):
    plt.subplot(2, 2, i+1)
    for t in range(Times):
        plt.scatter(result[i, t, 0][0], result[i, t, 0][1], color='black', marker='o')
        plt.scatter(result[i, t, 1][0], result[i, t, 1][1], color='red', marker='^')
    plt.title(title_list[i])
    plt.ylim(-2, 2)
    plt.xlim(-2, 2)
        
        
```

# Figure 2

```
t1 = time.time()


N = 2
df = 5
Times = 20
dim = [2, 20, 200, 2000]
N_dim = len(dim)
index = [i for i in range(N_dim)]


result = np.zeros((N_dim, Times, N, N))

for p, pc in zip(dim, index):
    print('Dimension : {}'.format(p))
    print('Calcurating...')
    
    for t in range(Times):
        X = np.random.normal(0, 1, (N, p)) / np.sqrt(np.random.chisquare(df, N) / df).reshape(N, -1)
        Sd = np.dot(X, X.T) / N
        
        value, vec = la.eig(Sd)
        for n in range(N):
            result[pc, t, n] = float(N) * value[n] * vec.T[n] / float(p)
        
       # result[pc, t] = float(N) / float(p) * Sd
        
    print('Result')
    print(result[pc])
    print('Finished')
    print('')

t2 = time.time() - t1
print('Calcuration time : {}'.format(t2))
```

```
plt.figure(figsize=(10, 10))
title_list = ['dim = {}'.format(p) for p in dim]
for i in range(N_dim):
    plt.subplot(2, 2, i+1)
    for t in range(Times):
        plt.scatter(result[i, t, 0][0], result[i, t, 0][1], color='black', marker='o')
        plt.scatter(result[i, t, 1][0], result[i, t, 1][1], color='red', marker='^')
    plt.title(title_list[i])
    plt.ylim(-4, 4)
    plt.xlim(-4, 4)
        
        
```
